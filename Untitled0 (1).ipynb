```python
import pandas as pd
link1='https://raw.githubusercontent.com/Myyn2304/pc3/refs/heads/main/Education%20expenditures.csv'
tabla1=pd.read_csv(link1)
link2='https://raw.githubusercontent.com/Myyn2304/pc3/refs/heads/main/Electricity%20-%20installed%20generating%20capacity.csv'
tabla2=pd.read_csv(link2)
link3='https://raw.githubusercontent.com/Myyn2304/pc3/refs/heads/main/Energy%20consumption%20per%20capita.csv'
tabla3=pd.read_csv(link3)
link4='https://raw.githubusercontent.com/Myyn2304/pc3/refs/heads/main/Reserves%20of%20foreign%20exchange%20and%20gold.csv'
tabla4=pd.read_csv(link4)
```

```python
# se observa que el unico problema en la tabla 2 columna kW son las ","
pd.unique(tabla2.Kw[~tabla2.Kw.astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
# se observa que el unico problema en la tabla 3 columna Btu/person son las ","

pd.unique(tabla3['Btu/Person'][~tabla3['Btu/Person'].astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
# se observa que el unico problema en la tabla 4 columna kW son las "," y el "$"
pd.unique(tabla4.Value[~tabla4.Value.astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
#Eliminando las "," en las columnas requeridas. En la 4 tambi√©n se elimina "$"
tabla2['Kw'] = tabla2['Kw'].astype(str).str.strip().str.replace(r',', '', regex=True)
tabla3['Btu/Person'] = tabla3['Btu/Person'].astype(str).str.strip().str.replace(r',', '', regex=True)
tabla4['Value'] = tabla4['Value'].astype(str).str.strip().str.replace(r',', '', regex=True).str.replace(r'^\$', '', regex=True)
```

```python
#verificando que ya no hay caracteres que no sean n√∫meros (ni "," ni "$")
pd.unique(tabla2.Kw[~tabla2.Kw.astype(str).str.contains(r'\d+\.\d+', regex=True)])
pd.unique(tabla3['Btu/Person'][~tabla3['Btu/Person'].astype(str).str.contains(r'\d+\.\d+', regex=True)])
pd.unique(tabla4.Value[~tabla4.Value.astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
#transformar todos los valores a numeros en las columnas requeridas
tabla2.Kw=pd.to_numeric(tabla2.Kw,errors='coerce')
#comprobar que todo en Kw es n√∫mero y hay 211 valores no nulos, es decir, que no hay NaN
tabla2.info()
```

```python
#transformar todos los valores a numeros en las columnas requeridas
tabla3['Btu/Person']=pd.to_numeric(tabla3['Btu/Person'],errors='coerce')
#comprobar que todo en Btu/person es n√∫mero y hay 195 valores no nulos, es decir, que no hay NaN
tabla3.info()
```

```python
#Iniciar en orden partiendo de %gdp, creando 5 intervalos
import mapclassify
import numpy as np
K=5
#Se usar√° FisherJenks al ser el m√°s eficiente
intervaloGDP=mapclassify.FisherJenks(tabla1234['%_of_GDP'], k=K)
intervalokW=mapclassify.FisherJenks(tabla1234['kW'], k=K)
intervaloBTU=mapclassify.FisherJenks(tabla1234['Btu/Person'], k=K)
intervaloVALUE=mapclassify.FisherJenks(tabla1234['Value'], k=K)
```

```python
#Se agregan los intervalos creados como columna
tabla1234['Intervalo_GDP']=intervaloGDP.yb
tabla1234['Intervalo_kW']=intervalokW.yb
tabla1234['Intervalo_BTU']=intervaloBTU.yb
tabla1234['Intervalo_VALUE']=intervaloVALUE.yb
tabla1234
```

```python
#quitando la columna slug pues ya no ser√° de uso
tabla1234.drop(['Slug'], axis=1, inplace=True)
```

```python
#reacomodando la tabla
reordenar=['Country','Region','%_of_GDP','Intervalo_GDP','kW','Intervalo_kW','Btu/Person','Intervalo_BTU','Value','Intervalo_VALUE']
tabla1234=tabla1234.reindex(columns=reordenar)
tabla1234
```
